```python
# pip install transformers accelerate torch pandas soundfile  # if needed

import os
import unicodedata as ud
import pandas as pd
import torch
from transformers import pipeline

# ---- CONFIG ----
CSV_PATH = "path/to/your_dataset.csv"          # the CSV you showed
AUDIO_DIR = "/content/drive/MyDrive/Ehugbo"                     # folder that contains the wav files
OUTPUT_CSV = "transcriptions_out.csv"
FILENAME_COL = "Wav file name"                 # from your screenshot
TEXT_COL = "Transcription1"                    # ground truth column (optional)

# Load the ASR model (plain Transformers)
device = 0 if torch.cuda.is_available() else -1
asr = pipeline(
    "automatic-speech-recognition",
    model="deepdml/wav2vec2-large-mms-1b-igbo-mix",
    device=device,
    # If your audios can be long, these help chunk them transparently:
    chunk_length_s=30,
    stride_length_s=(4, 2),
)

# Helper: build a valid .wav path; your CSV shows names like "John 1"
def make_wav_path(name: str) -> str:
    name = name.strip()
    # Add .wav if missing
    if not name.lower().endswith(".wav"):
        name = f"{name}.wav"
    return os.path.join(AUDIO_DIR, name)


# Load data
# df = pd.read_excel("/content/drive/MyDrive/Ehugbo/Ehugbo break'.xlsx")

# Build absolute paths
# df["audio_path"] = df[FILENAME_COL].apply(make_wav_path)

# (Optional) check missing files early
missing = df[~df["audio_path"].apply(os.path.exists)]
if len(missing):
    print("Warning: these files were not found:")
    print(missing[["audio_path"]].head(20))
    # You can decide to raise here if you want:
    # raise FileNotFoundError("Some audio files are missing.")
```

```python
# Transcribe
preds = []
for i,p in enumerate(df["audio_path"]):
    try:
        out = asr(p)              # {'text': '...'}
        preds.append(out["text"])
        print(f"Done for {i}")
    except Exception as e:
        preds.append("")          # or None
        print(f"Failed on {p}: {e}")

df["prediction"] = preds
```

```python
# pip install soundfile  # if you don't have it
import soundfile as sf
import numpy as np

def duration_seconds(path: str):
    try:
        info = sf.info(path)                   # reads header only (fast)
        return info.frames / info.samplerate   # seconds (float)
    except Exception:
        return np.nan                          # missing/bad file -> NaN

# If you already have df["audio_path"] from earlier:
df["seconds"] = df["audio_path"].apply(duration_seconds).round(3)

# If you only have the filename column, build paths first:
# df["audio_path"] = df["Wav file name"].apply(lambda n: os.path.join(AUDIO_DIR, f"{n}.wav" if not n.lower().endswith(".wav") else n))

# (Optional) integer “second count”
# df["seconds_int"] = df["seconds"].round().astype("Int64")
```

```python
import re
import unicodedata as ud
import pandas as pd

# ---------- normalization ----------
def nfc(s: str) -> str:
    return ud.normalize("NFC", s) if isinstance(s, str) else ""

def normalize_text(s: str) -> str:
    s = nfc(s).casefold()
    s = re.sub(r"[^\w\s’'-]", " ", s)  # keep letters/diacritics/'/-
    s = re.sub(r"[_\d]+", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

def strip_diacritics(s: str) -> str:
    return "".join(c for c in ud.normalize("NFD", s) if ud.category(c) != "Mn")

# ---------- Levenshtein (token-level) with counts ----------
def wer_counts(truth: str, hyp: str):
    t = normalize_text(truth).split()
    h = normalize_text(hyp).split()
    n, m = len(t), len(h)
    if n == 0:
        # all insertions if hyp has tokens; WER defined as 1.0 unless both empty
        return (0.0 if m == 0 else 1.0), 0, 0, m

    # DP table
    dp = [[0]*(m+1) for _ in range(n+1)]
    for i in range(1, n+1): dp[i][0] = i
    for j in range(1, m+1): dp[0][j] = j

    for i in range(1, n+1):
        ti = t[i-1]
        for j in range(1, m+1):
            if ti == h[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                dp[i][j] = 1 + min(dp[i-1][j],     # delete
                                   dp[i][j-1],     # insert
                                   dp[i-1][j-1])   # substitute

    # backtrace to count S/D/I
    i, j = n, m
    S = D = I = 0
    while i > 0 or j > 0:
        if i > 0 and j > 0 and t[i-1] == h[j-1]:
            i -= 1; j -= 1
        elif i > 0 and j > 0 and dp[i][j] == dp[i-1][j-1] + 1:
            S += 1; i -= 1; j -= 1
        elif i > 0 and dp[i][j] == dp[i-1][j] + 1:
            D += 1; i -= 1
        else:
            I += 1; j -= 1

    wer = (S + D + I) / n
    return wer, S, D, I

# ---------- Levenshtein (char-level) distance -> CER ----------
def cer_one(truth: str, hyp: str) -> float:
    t = normalize_text(truth)
    h = normalize_text(hyp)
    n, m = len(t), len(h)
    if n == 0:
        return 0.0 if m == 0 else 1.0

    prev = list(range(m+1))
    for i in range(1, n+1):
        cur = [i] + [0]*m
        for j in range(1, m+1):
            if t[i-1] == h[j-1]:
                cur[j] = prev[j-1]
            else:
                cur[j] = 1 + min(prev[j], cur[j-1], prev[j-1])
        prev = cur
    dist = prev[m]
    return dist / n

def wer_no_diacritics(truth: str, hyp: str) -> float:
    return wer_counts(strip_diacritics(truth), strip_diacritics(hyp))[0]

def cer_no_diacritics(truth: str, hyp: str) -> float:
    return cer_one(strip_diacritics(truth), strip_diacritics(hyp))

# ---------- apply to your dataframe (df has Transcription1, prediction, seconds) ----------
rows = []
for r, h in zip(df["Transcription1"], df["prediction"]):
    w, s, d, i = wer_counts(r, h)
    rows.append((w, s, d, i, cer_one(r, h), wer_no_diacritics(r, h), cer_no_diacritics(r, h)))

df[["wer", "subs", "dels", "ins", "cer", "wer_no_diac", "cer_no_diac"]] = rows
df["ser"] = (df["wer"] > 0).astype(int)

# corpus summary
print("WER:", df["wer"].mean())
print("CER:", df["cer"].mean())
# print("SER:", df["ser"].mean())
# print("Substitutions:", df["subs"].sum(), "Deletions:", df["dels"].sum(), "Insertions:", df["ins"].sum())
print("WER (no diacritics):", df["wer_no_diac"].mean(), "CER (no diacritics):", df["cer_no_diac"].mean())
```
